```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
year = np.array([1972, 1982, 1992, 2002, 2012])
e_india = np.array([100.6, 158.61, 305.54, 394.96, 724.79])
e_bang = np.array([10.5, 25.21, 58.65, 119.27, 274.87])
plt.xlabel("year")
plt.ylabel("Electricity consumption")
plt.title("Electricity Consumption in two countries")
# Exercise 2
plt.xlim((1980, 2000))
plt.ylim((0, 500))
plt.xticks([1985, 1990, 1995], ["x1", "x2", "x3"])
plt.yticks([0, 100, 200, 300, 400], ["y1", "y2", "y3", "y4", "y5"])
# Back to exercise 1
plt.plot(year, e_india, color = "r", marker = "o", label = "India")
plt.plot(year, e_bang, linestyle = "-.", color = "m", marker = "*", label = "Bangladesh")
plt.legend()
plt.show()
```
